## Checking the Multicolinearity 

```python
x = df_new.drop('fraud_reported',axis = 1)
y = df_new['fraud_reported']
z = x.columns
print(x.shape)
print(y.shape)
```

### Applying Standard Scaler

```python
#Scaling the data using standardscaler
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

x = sc.fit_transform(x)
x = pd.DataFrame(x,columns = z)
x
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def vif(x):
    vif = pd.DataFrame()
    vif['Column Name'] = x.columns
    vif['VIF Factor'] = [variance_inflation_factor(x.values,i) for i in range(x.shape[1])]
    return(vif)
```

```python
vif(x)
```

#### - As observed , the columns 'total_claim_amount', 'vehicle_claim', 'property_claim', 'injury_claim', 'age',and 'months_as_customer' are having high inflation variance. So let's remove columns 'total_claim_amount', 'months_as_customer' and try calculating the VIF.

```python
x1 = x.drop(['total_claim_amount','months_as_customer'], axis = 1)
vif(x1)
```

#### - Now the variance of inflation is reduced and it is below 5 for all the variables which indicates that there will be less multi collinearity in our data .

```python
lr = LogisticRegression()
knn = KNeighborsClassifier()
svc = SVC()
dtc = DecisionTreeClassifier()
gnb = GaussianNB()
rfc = RandomForestClassifier()
abc = AdaBoostClassifier()
gbc = GradientBoostingClassifier()
etc = ExtraTreesClassifier()
xgbc = XGBClassifier()
```

#### FINDING BEST RANDOM STATE-

```python
maxAcc = 0
maxRS = 0
for i in range(1,100):
    x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size =0.20, random_state = i)
   
    rfc.fit(x_train,y_train)
    pred = rfc.predict(x_test)
    acc  = accuracy_score(y_test,pred)
    if acc>maxAcc:
        maxAcc=acc
        maxRS=i
print(f"Best Accuracy is: {maxAcc} on random_state: {maxRS}")
```

```python
#Applying the best selected random state -

x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size =0.20, random_state = 76)
```

```python
def auc_roc(model):
    y_pred_prob = model.predict_proba(x_test)[:,1]
    fpr,tpr,thresholds = roc_curve(y_test,y_pred_prob)
    auc_score = roc_auc_score(y_test, model.predict(x_test))
    plt.plot([0,1],[0,1],'k--')
    plt.plot(fpr, tpr, label = 'AUC ='+str(round(auc_score,2)))
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend(loc = 'lower right')
    plt.title(model)
    plt.show()
```
