## DATA CLEANING - REMOVING SKEWNESS AND OUTLIERS

```python
### STEP 1-Splitting the data into 'x' and 'y' labels-

x = cen.drop('Income',axis = 1)
y = cen['Income']
print(x.shape)
print(y.shape)
```

```python
sns.countplot(y)
print(y.value_counts())
```

#### - There is a huge imbalance in the target data. The data for people below or equal to 50k are more in the dataset.so , we need to balance tha dataset.

```python
# Using SMOTE to balance the categorical data-

from imblearn.over_sampling import SMOTE    # Importing library
sm= SMOTE()
x,y=sm.fit_resample(x,y)
y
```

```python
#Rechecking the Data after using SMOTE - 

sns.countplot(y)
```

#### - Hence , we have sucessfully balanced the data of our target variable.

```python
print(x.shape)
print(y.shape)
```

```python
### STEP-2 Removing skewness 

#Removing skewness from the data with standard limit as +/-0.5. -

x.skew().sort_values(ascending=False)[np.abs(x.skew())>0.5]
```

#### -These are the variables which are having high skewness which is beyond the standard limit.

```python
# Importing library to remove skewness .

from sklearn.preprocessing import power_transform

def skew(col):
    x[col]=power_transform((x[col].values.reshape(-1,1)))

skew("Capital_gain")
skew("Capital_loss")
skew("Fnlwgt")
skew("Relationship")
skew("Education")
skew("Race")
skew("Native_country")
```

```python
x.skew().sort_values(ascending = False)
```

#### - As the skewness is still high for the columns 'Capital_loss'and 'Capital_gain', we can drop these columns from the dataset.

```python
x.drop(['Capital_loss','Capital_gain'],axis = 1, inplace = True)
x.skew().sort_values(ascending = False)
```

#### - Finally , skewness is reduced.

```python
## Step -3 Removing Outliers using z score method 
from scipy.stats import zscore
df = x.copy()
df['Income'] = y.copy()
df.drop(['Workclass', 'Education', 'Race', 'Native_country','Income'],axis = 1, inplace = True)    #removing the categorical column from the dataset before removing outliers.
z = np.abs(zscore(df))
z
```

```python
threshold = 3
df_new = df[(z<3).all(axis=1)]
df_new['Workclass'] = x['Workclass']
df_new['Education'] = x['Education']
df_new['Race'] = x['Race']
df_new['Native_country'] = x['Native_country']
df_new['Income'] = y.copy()
df_new
```

```python
df['Workclass'] = x['Workclass']
df['Education'] = x['Education']
df['Race'] = x['Race']
df['Native_country'] = x['Native_country']
df['Income'] = y.copy()
df.shape
```

```python
# SPLITTING THE DATA INTO X AND Y VARIABLE.

X = x.copy()
Y = y.copy()
```

#### FINDING BEST RANDOM STATE-

```python
maxAcc = 0
maxRS = 0
for i in range(1,100):
    x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size =0.20, random_state = i)
   
    rfc.fit(x_train,y_train)
    pred = rfc.predict(x_test)
    acc  = accuracy_score(y_test,pred)
    if acc>maxAcc:
        maxAcc=acc
        maxRS=i
print(f"Best Accuracy is: {maxAcc} on random_state: {maxRS}")
```

```python
#Applying the best selected random state -

x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size =0.20, random_state = 96 )
```

```python
# Training all the models together-

rfc.fit(x_train,y_train.ravel())
lr.fit(x_train,y_train.ravel())
knn.fit(x_train,y_train.ravel())
svc.fit(x_train,y_train.ravel())
dtc.fit(x_train,y_train.ravel())
abc.fit(x_train,y_train.ravel())
gbc.fit(x_train,y_train.ravel())
gnb.fit(x_train,y_train.ravel())
etc.fit(x_train,y_train.ravel())
xgbc.fit(x_train,y_train.ravel())
```

```python
def auc_roc(model):
    y_pred_prob = model.predict_proba(x_test)[:,1]
    fpr,tpr,thresholds = roc_curve(y_test,y_pred_prob)
    auc_score = roc_auc_score(y_test, model.predict(x_test))
    plt.plot([0,1],[0,1],'k--')
    plt.plot(fpr, tpr, label = 'AUC ='+str(round(auc_score,2)))
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend(loc = 'lower right')
    plt.title(model)
    plt.show()
```
